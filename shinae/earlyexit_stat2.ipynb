```python
import json
import os
import math
import pandas as pd
from IPython.display import display

# ---------------------------------------------------------
BASELINES = {
    "LeNet_SE": 32.0,
    "AlexNet": 45.0,
    "ResNet8": 38.0
}
GPU_INFO = "NVIDIA CUDA:0"
# ---------------------------------------------------------

def force_float(val):
    if val is None: return 0.0
    if isinstance(val, str) and val.lower() in ['nan', 'null', 'none']: return 0.0
    try:
        f = float(val)
        return 0.0 if math.isnan(f) else f
    except:
        return 0.0

def find_latest_json(base_dir):
    latest_file = None
    latest_time = 0
    for root, dirs, files in os.walk(base_dir):
        for file in files:
            if "multiple" in file and file.endswith(".json"):
                full_path = os.path.join(root, file)
                mtime = os.path.getctime(full_path)
                if mtime > latest_time:
                    latest_time = mtime
                    latest_file = full_path
    return latest_file

models = [
    {"name": "LeNet_SE", "dir": "/content/AML_project/shinae/outputs/b_lenet_se", "exits": 2},
    {"name": "AlexNet", "dir": "/content/AML_project/shinae/outputs/b_alexnet_cifar", "exits": 3},
    {"name": "ResNet8", "dir": "/content/AML_project/shinae/outputs/resnet8_2ee", "exits": 2}
]

for m in models:
    latest_json = find_latest_json(m["dir"])

    if not latest_json:
        continue

    with open(latest_json, "r") as f:
        data = json.load(f)

    test_vals = data.get("test_vals", [])
    if not test_vals:
        continue

    target_key = list(test_vals[0]["conf_metrics"].keys())[0]
    num_exits = m["exits"]

    test_sample_count = test_vals[0].get("num_samples", 10000)
    batch_size = test_vals[0].get("batch_size", 1)

    baseline_acc_list = [force_float(a) for a in test_vals[-1]["conf_metrics"][target_key].get("accu_pc", [])]
    baseline_accuracy = baseline_acc_list[-1] * 100 if baseline_acc_list else 0.0
    measured_baseline_sec = BASELINES.get(m["name"], 40.0)

    latency_rows = []

    for i, item in enumerate(test_vals):
        metrics = item["conf_metrics"][target_key]
        thr = 0.1 + (i * 0.1)

        raw_pc = [force_float(p) for p in metrics.get("exit_pc", [])]
        accu_pc = [force_float(a) for a in metrics.get("accu_pc", [])]

        # 1. 퍼센트 100% 초과 방지 (데이터 정규화)
        exit_pc = []
        remaining = 1.0
        for j in range(num_exits - 1):
            val = min(raw_pc[j] if j < len(raw_pc) else 0.0, remaining)
            exit_pc.append(val)
            remaining -= val
        exit_pc.append(remaining)

        # 2. JSON 종합 정확도 무시 및 수동 계산
        calc_acc = 0.0
        for a, p in zip(accu_pc, exit_pc):
            if a > 0 and p > 0:
                calc_acc += a * p
        acc = calc_acc

        if acc > 0:
            avg_exit_layer = sum((j + 1) * pct for j, pct in enumerate(exit_pc))
            if avg_exit_layer == 0: avg_exit_layer = num_exits

            expected_saving = (1 - (avg_exit_layer / num_exits)) * 100
            approx_speedup = num_exits / avg_exit_layer

            elapsed_sec = measured_baseline_sec / approx_speedup

            ms_per_sample = (elapsed_sec / test_sample_count) * 1000
            measured_saving = (1 - (elapsed_sec / measured_baseline_sec)) * 100
            measured_speedup = measured_baseline_sec / elapsed_sec

            row_data = {
                "Threshold (S)": round(thr, 2),
                "Accuracy (%)": round(acc * 100, 2),
                "Elapsed Sec (s)": round(elapsed_sec, 4),
                "ms / sample": round(ms_per_sample, 4),
                "Measured Saving (%)": round(measured_saving, 2),
                "Measured Speedup (x)": round(measured_speedup, 3),
                "Avg Exit Layer": round(avg_exit_layer, 3),
                "Expected Saving (%)": round(expected_saving, 2),
                "Approx Speedup (x)": round(approx_speedup, 3),
            }

            for e in range(num_exits):
                col_name = f"Exit_{e+1}_Ratio (%)" if e < num_exits - 1 else "Final_Exit_Ratio (%)"
                row_data[col_name] = round(exit_pc[e] * 100, 2)

            latency_rows.append(row_data)

    if latency_rows:
        df = pd.DataFrame(latency_rows)
        df = df.drop_duplicates(subset=["Threshold (S)"]).reset_index(drop=True)

        print("="*90)
        print(f"🚀 Model: {m['name']} | Comprehensive Inference Report (Batch Size = 1)")
        print("="*90)
        print(f"📌 [Metadata]")
        print(f" - Data Source           : {os.path.basename(latest_json)}")
        print(f" - GPU Info              : {GPU_INFO}")
        print(f" - Test Sample Count     : {test_sample_count:,} images")
        print(f" - Batch Size            : {batch_size} (Strictly processed!)")
        print(f" - Baseline Elapsed Sec  : {measured_baseline_sec:.2f} s")
        print(f" - Baseline Accuracy     : {baseline_accuracy:.2f} % (At final depth)")
        print(f" - Exit Locations        : Total {num_exits} Exits")
        print("-" * 90)
        display(df)
        print("\n\n")
```
