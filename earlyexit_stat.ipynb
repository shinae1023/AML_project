```python
import json
import glob
import os
import math
import pandas as pd
from IPython.display import display

def force_float(val):
    if val is None: return 0.0
    if isinstance(val, str) and val.lower() in ['nan', 'null', 'none']: return 0.0
    try:
        f = float(val)
        return 0.0 if math.isnan(f) else f
    except:
        return 0.0

# 변경된 경로 반영
models = [
    {"name": "LeNet (MNIST)", "path": "/content/AML_project/shinae/outputs/b_lenet_se/range_notes/*multiple*.json", "exits": 2},
    {"name": "AlexNet (CIFAR-10)", "path": "/content/AML_project/shinae/outputs/b_alexnet_cifar/range_notes/*multiple*.json", "exits": 3},
    {"name": "ResNet8 (CIFAR-10)", "path": "/content/AML_project/shinae/outputs/resnet8_2ee/range_notes/*multiple*.json", "exits": 2}
]

for m in models:
    json_files = glob.glob(m["path"])
    if not json_files:
        print(f"⚠️ {m['name']}의 JSON 스캔 파일을 찾을 수 없음. 경로를 확인하셈.")
        continue

    latest_json = max(json_files, key=os.path.getctime)

    with open(latest_json, "r") as f:
        data = json.load(f)

    test_vals = data.get("test_vals", [])
    if not test_vals: continue

    target_key = list(test_vals[0]["conf_metrics"].keys())[0]
    num_exits = m["exits"]
    latency_rows = []

    for i, item in enumerate(test_vals):
        metrics = item["conf_metrics"][target_key]

        # 주입된 인덱스를 바탕으로 올바른 Threshold 역산
        thr = 0.1 + (i * 0.1)

        exit_pc = [force_float(p) for p in metrics.get("exit_pc", [])]
        acc = force_float(metrics.get("combined_accuracy"))

        # NaN 결측치 수동 복구 연산
        if acc == 0.0:
            acc_list = metrics.get("accu_pc", [])
            calc_acc = 0.0
            for a, p in zip(acc_list, exit_pc):
                if force_float(p) > 0 and force_float(a) > 0:
                    calc_acc += force_float(a) * force_float(p)
            acc = calc_acc

        if acc > 0:
            avg_exit_layer = sum((j + 1) * pct for j, pct in enumerate(exit_pc))
            if avg_exit_layer == 0: avg_exit_layer = num_exits

            expected_saving = (1 - (avg_exit_layer / num_exits)) * 100
            approx_speedup = num_exits / avg_exit_layer

            latency_rows.append({
                "Threshold (S)": round(thr, 2),
                "Accuracy (%)": round(acc * 100, 2),
                "EE1 Ratio (%)": round(exit_pc[0] * 100, 2),
                "Avg Exit Layer": round(avg_exit_layer, 3),
                "Expected Saving (%)": round(expected_saving, 2),
                "Approx Speedup (x)": round(approx_speedup, 2)
            })

    if latency_rows:
        df = pd.DataFrame(latency_rows)
        df = df.drop_duplicates(subset=["Threshold (S)"]).reset_index(drop=True)

        print(f"\n📊 [출력 완료] {m['name']} 성능 통합 지표 테이블")
        display(df)
```
